```python
countries_df = pd.DataFrame()
```

```python
for i in df['country'].unique():
    print(i, ':', len(df[df['country']==i]))
    countries_df.loc[i, 'count'] = len(df[df['country']==i])
```

```python
sum(countries_df.sort_values(by='count', ascending=False).head(50)['count'])
```

```python
sum(countries_df.sort_values(by='count', ascending=False)[50:100]['count'])
```

```python
for i in df['country'].unique():
    countries_df.loc[i, 'target_mean'] = np.mean(df[df['country']==i]['TARGET'])
    countries_df.loc[i, 'target_std'] = np.std(df[df['country']==i]['TARGET'])
```
