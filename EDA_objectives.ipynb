```python
countries_df = pd.DataFrame()
```

```python
for i in df['country'].unique():
    countries_df.loc[i, 'count'] = len(df[df['country']==i])
    countries_df.loc[i, 'target_mean'] = np.mean(df[df['country']==i]['TARGET'])
    countries_df.loc[i, 'target_std'] = np.std(df[df['country']==i]['TARGET'])
```

```python
continental_df = pd.DataFrame()
for i in df['subcontinent'].unique():
    continental_df.loc[i, 'subcontinental'] = len(df[df['subcontinent']==i])
    continental_df.loc[i, 'target_mean'] = np.mean(df[df['subcontinent']==i]['TARGET'])
    continental_df.loc[i, 'target_std'] = np.std(df[df['subcontinent']==i]['TARGET'])
```
